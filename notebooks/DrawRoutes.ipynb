```python
valid_routes = ['odometry_data_1666784805698008326.csv_si.csv',
                'odometry_data_1666780469251254852.csv_si.csv',
                'odometry_data_1666781426581501015.csv_si.csv',
                'odometry_data_1666781058545170756.csv_si.csv',
                'odometry_data_1666783210242100442.csv_si.csv',
                'odometry_data_1666784037325996273.csv_si.csv',
]
subjects = ['Subj1', 'Subj2','Subj3','Driver1','Driver2','Driver3']
```

```python
reference = pd.read_csv(os.path.join(route_dir, 'reference.csv'))
reference['x'] += 1
reference['y'] += 1
reference = reference.round(2)

reference = reference.drop_duplicates()


for i, route in enumerate(valid_routes):
    route_data = pd.read_csv(os.path.join(route_dir, route))
    
    route_data = route_data[['t(s)', 'x(m)', 'y(m)']]
    #print(route_data.head())
    route_data = route_data.round(2)

    route_data = route_data.drop_duplicates()

    multiplier = int((len(route_data)) / len(reference))  
    try:
        route_data = route_data.iloc[::multiplier, :]
    except:
        continue

    while len(route_data)!= len(reference):
        route_data = route_data.drop(route_data.index[np.random.randint(0, len(route_data))])
    

    plt.subplot(2, 3, i+1)
    X_route_np = np.array(route_data['x(m)'])
    Y_route_np = np.array(route_data['y(m)'])

    X_ref_np = np.array(reference['x'])
    Y_ref_np = np.array(reference['y'])
    
    X_diff = np.subtract(X_route_np, X_ref_np)
    Y_diff = np.subtract(Y_route_np, Y_ref_np)

    diff = np.sqrt(X_diff**2 + Y_diff**2)
    
    e = round(np.sum(diff),2)
    
    plt.title(f'{subjects[i]}: {e}')
    plt.plot(route_data['x(m)'], route_data['y(m)'])
    plt.plot(reference['x'], reference['y'])
    plt.xticks([])
    plt.yticks([])
plt.show()

       



plt.savefig('route_plot.jpg')
plt.show()
```
